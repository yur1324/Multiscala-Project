```python
#@title Solvation (adding water)
#CADY
!gmx solvate -cp pdb_nostruct/cady_1box.gro -cs spc216.gro -o pdb_nostruct/cady_1solvated.gro -p pdb_nostruct/cady_1.top 
```

```python
#MAP
!gmx solvate -cp pdb_struct/map_1box.gro -cs spc216.gro -o pdb_struct/map_1solvated.gro -p pdb_struct/map_1.top 
```

```python
# Quick peek cady:
import nglview as nv
with open("pdb_nostruct/cady_1solvated.gro") as f:
    view = nv.show_file(f, ext="gro")
view.add_ball_and_stick(selection="water")
view.camera = 'orthographic'
view
```

```python
# Quick peek map:
import nglview as nv
with open("pdb_struct/map_1solvated.gro") as f:
    view = nv.show_file(f, ext="gro")
view.add_ball_and_stick(selection="water")
view.camera = 'orthographic'
view
```

```python
#@title Neutralization (adding ions Na+ and Cl-)
#CADY
!touch pdb_nostruct/nostructdummy.mdp #creazione file
!gmx grompp -f pdb_nostruct/nostructdummy.mdp -c pdb_nostruct/cady_1solvated.gro -p pdb_nostruct/cady_1.top -o pdb_nostruct/cady_1ions.tpr #"intermedio", finchè dummy.mdp è vuoto
!echo "SOL" | gmx genion -s pdb_nostruct/cady_1ions.tpr -p pdb_nostruct/cady_1.top -o pdb_nostruct/cady_1box_ions.gro -conc 0.15 -neutral #concentrazione 0.15M.
with open("pdb_nostruct/cady_1box_ions.gro") as f:
    view = nv.show_file(f, ext="gro")
view.add_licorice(selection="water")
view.add_ball_and_stick(selection="NA",color='blue',aspectRatio='5') 
view.add_ball_and_stick(selection="CL",color='green',aspectRatio='5')
view.camera = 'orthographic'
view
```

```python
#MAP
!touch pdb_struct/structdummy.mdp #creazione file
!gmx grompp -f pdb_struct/structdummy.mdp -c pdb_struct/map_1solvated.gro -p pdb_struct/map_1.top -o pdb_struct/map_1ions.tpr #"intermedio", finchè dummy.mdp è vuoto
!echo "SOL" | gmx genion -s pdb_struct/map_1ions.tpr -p pdb_struct/map_1.top -o pdb_struct/map_1box_ions.gro -conc 0.15 -neutral #concentrazione 0.15M.
with open("pdb_struct/map_1box_ions.gro") as f:
    view = nv.show_file(f, ext="gro")
view.add_licorice(selection="water")
view.add_ball_and_stick(selection="NA",color='blue',aspectRatio='5') 
view.add_ball_and_stick(selection="CL",color='green',aspectRatio='5')
view.camera = 'orthographic'
view
```
